```python
#Import necessary libraries
import numpy as np
import pandas as pd
```

```python
#Load the dataset
data=pd.read_csv("/content/Country_clusters.csv")
data.head()
```

```python
# Select the relevant columns for clustering
new_data=data.iloc[:,1:3]
new_data.head()
```

```python
# Import KMeans algorithm for clustering
from sklearn.cluster import KMeans
# Initialize the KMeans model by setting the number of clusters to 3
kmeans=KMeans(n_clusters=3)
```

```python
# Train the model
kmeans.fit(new_data)
```

```python
# Predict the clusters for each data point
clusters=kmeans.fit_predict(new_data)
```

```python
# Print the cluster labels (0, 1, 2) for each data point
print(clusters)
```
